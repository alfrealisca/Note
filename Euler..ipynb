```julia
function euler(t0, x0, h, N, f)
    t_actual = t0
    x_actual = x0
    
    ts = Float64[]
    xs = Float64[]

    for i in t0:h:N
        push!(ts, t_actual)
        push!(xs, x_actual)
        
        t_nueva = t_actual + h
        x_nueva = x_actual + h*f(t_actual, x_actual)

        t_actual = t_nueva
        x_actual = x_nueva
    end
    plot(ts, xs)
end
```

```julia
α(t, x) = 5t
```

```julia
α(2,4)
```

```julia
euler(1,0,0.0625,100,α)
```
