```julia
function euler(t0, x0, h, N, f)
    t_attuale = t0
    x_attuale = x0
    
    ts = zeros(Float64, N)
    xs = zeros(Float64, N)
    
    for i in 1:N
        ts[i] = t_attuale
        xs[i] = x_attuale
        
        t_nuova = t_attuale + h
        x_nuova = x_attuale + h*f(t_attuale, x_attuale)
        
        t_attuale = t_nuova
        x_attuale = x_nuova
        
    end
    ts, xs
end
```

mo alcuni esempi:

```julia
using PyPlot
```

**Esempio 1.**

```julia
esempio1(t, x) = t^3
```

```julia
euler(0, 0, 0.125, 100, esempio1)
```

```julia
asse_t1, asse_x1 = euler(0, 0, 0.125, 100, esempio1)
plot(asse_t1, asse_x1)
title("Soluzione esempio1")
xlabel(L"t")
ylabel(L"x(t)");
```
